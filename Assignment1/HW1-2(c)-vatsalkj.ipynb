```python
# Problem 2 (c)

import numpy as np
def same_frequency(num1,num2):
    num1 = str(num1)
    num2 = str(num2)
    num1_list = []
    num2_list = []
    count1 = []
    count2 = []
    for i in range(len(num1)):
        num1_list.append(num1[i])
    for i in range(len(num2)):
        num2_list.append(num2[i])
    key1 = np.unique(num1_list)
    key2 = np.unique(num2_list)
    key1.sort()
    key2.sort()
    for i in range(len(key1)):
        c = num1_list.count(key1[i])
        count1.append(c)
    for i in range(len(key2)):
        c = num2_list.count(key2[i])
        count2.append(c)
    if np.array_equal(count1,count2) and np.array_equal(key1,key2):
        return True
    else:
        return False
print(same_frequency(551122,221515))
print(same_frequency(321142,3212215))
print(same_frequency(12345,31354))
print(same_frequency(1212, 2211))
```
