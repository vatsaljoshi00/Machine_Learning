```python
# Problem 1 (c)

import numpy as np
from numpy import loadtxt

List = np.loadtxt('data.csv', dtype= 'float', delimiter=',', skiprows= 1)
m,n = np.shape(List)
X = np.ones((m,2))
X[:,1] = List[:,0] 
y = np.ones((m,1))
y[:,0] = List[:,1]
A = np.matmul(X.transpose(),X)
A = np.linalg.inv(A)
theta = np.matmul(A,X.transpose())
theta = np.matmul(theta,y)
print(theta)
```
