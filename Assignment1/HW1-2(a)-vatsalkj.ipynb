```python
# Problem 2 (a) 

def includes(item,val,start_ind=None):
    if type(item) == list:
        if val in item[start_ind:]:
            return True
        else: 
            return False
    elif type(item) == dict:
        if val in item.values():
            return True
        else:
            return False
    elif type(item) == str:
        if val in item[start_ind:]:
            return True
        else:
            return False
print(includes([2, 3, 4], 2, 0))
print(includes([2, 3, 4], 4, 1))
print(includes({'a':1,'b':2}, 1))
print(includes({'a':1,'b':2}, 'a'))
print(includes('abcd', 'a'))
```
