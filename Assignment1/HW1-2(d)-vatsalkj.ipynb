```python
# Problem 2 (d)

import numpy as np
import matplotlib.pyplot as plt

b = np.load('kmeans.npz')
centres = b['centers.npy']
data = b['data.npy']
pred = b['pred.npy']
for i in range(len(pred)):
    if pred[i] == 0:
        plt.plot(data[i,0], data[i,1], 'bo')
    elif pred[i] == 1:
        plt.plot(data[i,0], data[i,1], 'go')
    elif pred[i] == 2:
        plt.plot(data[i,0], data[i,1], 'co')
    elif pred[i] == 3:
        plt.plot(data[i,0], data[i,1], 'mo')
plt.plot(centres[:,0], centres[:,1], 'ro') 
plt.show()
```
