```python
# Problem 2 (b)

global m, n, Avg
Avg = 0
n = 0
m = 0
def moving_average(val = None):
    global m, n, Avg
    if m == 0 and val == None:
        Avg = 0
        print(round(Avg, 1))
        n = 0
    elif val == None:
        m += 1
        print(round(Avg, 1))
    else:
        n += 1
        Avg = (Avg*(n-1) + val)/n
        print(round(Avg, 1))
        m += 1
mAvg = moving_average
mAvg(10)
mAvg(11)
mAvg(12)
```
