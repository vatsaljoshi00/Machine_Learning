```python
# Problem 1 (b)

import numpy as np
print("Input number of data points: ")
m = int(input())
print("Enter data points here: ")
List = np.zeros((m,2))
for i in range(m):
    t = input()
    a = tuple(int(x) for x in t.split())
    List[i,0] = a[0]
    List[i,1] = a[1]

X = np.ones((m,1))
x = np.ones((m,1))
x[:,0] = List[:,0] 
X = np.concatenate((X, x), axis = 1)
y = np.ones((m,1))
y[:,0] = List[:,1]
A = np.matmul(X.transpose(),X)
A = np.linalg.inv(A)
theta = np.matmul(A,X.transpose())
theta = np.matmul(theta,y)
print(theta)
# The values are equal when compared to the first part.

```
