```python
import numpy as np
import time
# First Part
np.random.seed(24787)
a = np.random.randint(8, size=(3,4,4))
print("A = ",a)
print("Shape of A = ",np.shape(a))

# Locate value = 4
res = np.where(a == 4)
coordinates = list(zip(res[0], res[1], res[2]))
for i in coordinates:
    print("(depth, row, column) = ",i)

# Second Part
b = np.tile(a, (1, 2, 2))

# Third Part
c = np.sum(b, axis=0)
print("C = ",c)
print("Shape of C = ", np.shape(c))
```

```python
# Fourth Part
np.random.seed(24787)
a = np.random.randint(8, size=(1000,1000))
b = np.random.randint(8, size=(1000,1000))


c = np.zeros(a.shape)
def matmul(a,b):
    for i in range(a.shape[0]):
        for j in range(b.shape[1]):
            c[i,j] = np.dot(a[i,:], b[:,j].transpose())

z1 = time.time()
matmul(a,b)
print("Time elapsed = ", time.time() - z1)
print("Using row wise multiplicaiton: ",c)

z2 = time.time()
d = a@b
print("Using '@' operator:", d)
print("Time elapsed = ", time.time() - z2)

compare = c == d
result = compare.all()
  
print(result)
```

```python
'''
The inbuilt @ function is multiple times faster because of its vectorized form of execution. The for loops take a long time as compared
to its corresponding vectorized form.
'''
```
