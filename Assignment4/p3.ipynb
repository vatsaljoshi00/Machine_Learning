```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import pandas as pd
import numpy as np

# Define Sequential model with 3 layers
model = keras.Sequential(
    [
        layers.Dense(512, activation="relu", name="layer1"),
        layers.Dense(256, activation="relu", name="layer2"),
        layers.Dense(20, activation="softmax", name="layer3"),
    ]
)
# Call model on a test input
X1_train = pd.read_csv('Final_Train_Data.csv',header=None).iloc[:,0:1].values
X1_train = X1_train.reshape(-1,1)
X2_train = pd.read_csv('Final_Train_Data.csv',header=None).iloc[:,1:2].values
X2_train = X2_train.reshape(-1,1)


X1_test = pd.read_csv('Final_Test_Data.csv',header=None).iloc[:,0:1].values
X1_test = X1_test.reshape(-1,1)

X2_test = pd.read_csv('Final_Test_Data.csv',header=None).iloc[:,1:2].values
X2_test = X2_test.reshape(-1,1)

X_train = np.concatenate((X1_train,X2_train),axis = 1)
X_test = np.concatenate((X1_test,X2_test),axis = 1)

Y_train = pd.read_csv('Final_Train_Data.csv',header=None).iloc[:,2].values
Y_test = pd.read_csv('Final_Test_Data.csv',header=None).iloc[:,2].values

y = model(X_train)

model.compile(
    optimizer=keras.optimizers.Adam(lr = 0.01),  # Optimizer
    loss=keras.losses.SparseCategoricalCrossentropy(),
    metrics=[keras.metrics.SparseCategoricalAccuracy()],
)

print("Fit model on training data")
history = model.fit(
    X_train,
    Y_train,
    batch_size=64,
    epochs=100,
    # We pass some validation for
    # monitoring validation loss and metrics
    # at the end of each epoch
    validation_data=(X_test, Y_test),
)

_,train_acc = model.evaluate(X_train,Y_train,verbose=0)
_,test_acc = model.evaluate(X_test,Y_test,verbose=0)
print("Train accuracy: ", train_acc)
print("Test accuracy: ", test_acc)
```
